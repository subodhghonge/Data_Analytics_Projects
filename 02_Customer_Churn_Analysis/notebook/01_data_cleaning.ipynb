```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/Raw/Customer Churn.csv")
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0") #Replaced Blank spaces with 0 as tenure is 0
df["TotalCharges"] = df["TotalCharges"].astype("float") #assign datatype as float
```

```python
df["customerID"].duplicated().sum() #Checking Duplicated using Unique ID i.e. CustomerID
```

```python
#Coverted 0 and 1 values of senior citizen to yes/no to make it easier to understand
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```
