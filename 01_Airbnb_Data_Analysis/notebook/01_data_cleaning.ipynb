```python
df["reviews_per_month"].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['id','host_name'], axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
# # Convert 'price' column to numeric (this will force non-numeric values to NaN)
# df['price'] = pd.to_numeric(df['price'], errors='coerce')

# # Handle missing values in 'price' (fill with median or drop rows)
# df['price'].fillna(df['price'].median(), inplace=True)
```

```python
#Removing Outliers from Price Column
price_threshold = df['price'].quantile(0.99)
df = df[df['price'] < price_threshold]

df.head()
```

```python
min_nights_threshold = df['minimum_nights'].quantile(0.99)
df = df[df['minimum_nights'] < min_nights_threshold]
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
print("\nData after Cleaning")
df.describe()
```

```python
df.to_csv("../data/processed/airbnb_cleaned.csv", index=False)
```
