```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/airbnb_cleaned.csv")
df.head(10)
```

```python
#Distribution of Prices
plt.Figure(figsize=(8,5))
sns.histplot(df["price"], bins=50, kde=True)
plt.title("Price Distribution of Airbnb Listings")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.savefig("../reports/visuals/total_bill_distribution.png", bbox_inches = "tight")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df["neighbourhood_group_cleansed"], y=df["price"], estimator=np.mean)
plt.title("Average Price by Neighborhood Group")
plt.xlabel("Neighborhood Group")
plt.ylabel("Average Price")
plt.savefig("../reports/visuals/average_price_neighbourhood.png", bbox_inches = "tight")
plt.show()
```

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix for numerical features
corr = numerical_df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.savefig("../reports/visuals/correlation_heatmap.png", bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x="minimum_nights", data=df, palette="Set2")
plt.title("Distribution of Minimum Nights")
plt.xlabel("Minimum Nights")
plt.ylabel("Count")
plt.savefig("../reports/visuals/Minimum_Nights_distributions.png", bbox_inches="tight")
plt.show()
```

```python
#Price Vs Number of reviews (scatter plot)
plt.figure(figsize=(8,5))
sns.scatterplot(x="number_of_reviews", y="price", data=df)
plt.title("Price Vs. Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylabel("Price")
plt.savefig("../reports/visuals/price_vs_reviews.png", bbox_inches="tight")
plt.show()
```

```python
#Price by room type
plt.Figure(figsize=(10,5))
sns.boxplot(x="room_type", y="price", data=df)
plt.title("Price Distribution by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price")
plt.savefig("../reports/visuals/price_by_room_type.png", bbox_inches="tight")
plt.show()
```
